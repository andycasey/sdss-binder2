# Welcome to BinderHub!
Hosted by the Flatiron Institute, a division of the Simons Foundation.

BinderHub makes it easy for collaborators to access and analyze data. The Flatiron Institute hosts some public releases of the Sloan Digital Sky Survey (SDSS), and various subsets of proprietary data from SDSS-V.

## Access

Access to the SDSS BinderHub hosted by the Flatiron Institute is limited to SDSS collaborators and institutes. Access is provisioned through Google authentication.

I have added all SDSS-V collaborators by their given email address, which was stored in the SDSS internal collaboration database. However, the email address listed for each person is usually their institutional email address, which may not be managed by Google. If you or a colleague needs a Google email address added to the list, send an email to Andy Casey (acasey@flatironinstitute.org) and the SDSS-V Director, Juna Kollmeier (jak@carnegiescience.edu).

## Data

Public and proprietary SDSS data are accessible through this BinderHub. You should be cognizant of which data are public, and which are private. If in doubt, ask someone (anyone)!

All data are mounted in **read-only mode**, so don't worry: there's no way you could accidentally delete it.

## Storage

BinderHub lets you drag-and-drop to upload or download files. It's very handy. However, **there is no ephemeral storage**.


You should only ever save things to your `home/` directory. If you save things anywhere else, **it will be deleted when your notebook server finishes**. Please try to keep things stored in your `home/` directory to be less than 1 TB.

Do not store any sensitive or critical data on this service. Anything you put in your <code>home/</code> directory may be made available to Flatiron Institute researchers for the purpose of collaboration. All storage and resources should be considered best-effort scratch space and come with no guarantees or backups. Your entire account, including files in your <code>home/</code> directory will be removed after 7 days of inactivity.

## Notebooks

Some notebooks are included to help get you started. These are copies of templates: you can run and edit them, but they won't affect other people's BinderHub instances.

If you make a notebook that you think is helpful or neat in any way, **please add it** to the [SDSS BinderHub repository](https://github.com/andycasey/sdss-binder/). If it is added to that repository, then it will be available to every collaborator when they login!

## Marimo Notebooks

Marimo notebooks are an evolved version of Jupyter notebooks. Jupyter notebooks can often have 'state issues' when cells are executed out of order. Marimo notebooks fix this by having an internal representation of which cells affect other cells. When a variable or code in one cell is updated, Marimo will automatically re-execute any affected cells. This makes the entire notebook 'live', allowing you to create applications and interactive analysis tools. 

Marimo notebooks are stored as pure Python code, so they are easier to edit and store than Jupyter notebooks. 

You can load Marimo notebooks through this BinderHub instance by navigating to Marimo from the Launcher page.
